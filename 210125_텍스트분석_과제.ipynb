감성분석이란, 문서의 주관적인 감성/의견/감정/기분 등을 파악하기 위한 방법으로 소셜 미디어. 여론조사, 온라인 리뷰, 피드백 등 다양한 분야에서 활용되고 있다.

* 감성분석 단계 : 데이터 수집 -> 주관성 탐지 -> 극성 탐지
